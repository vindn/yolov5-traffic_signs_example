```python
# Predicao de uma placa utilizando os pesos do treinamento anterior

import cv2
import torch
from PIL import Image

model = torch.hub.load('ultralytics/yolov5', 'custom', './runs/train/exp11/weights/best.pt')

img1 = Image.open('road197.png')  # PIL image
imgs = [img1]  # lista de imagens

# Inference
results = model(imgs, size=416)  # includes NMS

# Results
results.print()  
results.save()  # or .show()

results.xyxy[0]  # img1 predictions (tensor)
results.pandas().xyxy[0]  # img1 predictions (pandas)
```
